# **5. Core Data Structures**
**Series**

* One-dimensional labeled array

* Can hold any data type (int, float, string, object)
* Each element has a unique label called index
* Used to track changes/patterns over time (e.g., daily temperature, stock prices)

**DataFrame**

* Two-dimensional labeled data structure
* Like a database table, Excel sheet, or CSV file
* Contains rows and columns

# **7. Saving/Exporting Data**

```python
import pandas as pd

data = {
    "Name": ["Asha", "Rahul", "Neha"],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

# Save to CSV
df.to_csv("my_students.csv", index=True)
print (data)

# Save to excel
df.to_excel("my_students.xlsx", index=False)

# Save to json
df.to_json("my_students.json")

# Saving as Text File
df.to_csv("my_students.txt", sep="\t", index=False)
```


df.to_csv() → Function to save as CSV

"students.csv" → File name

index=False → Do not save row numbers

If you remove index=False, it will also save index column.

These "`my_students.csv` , `my_students.xlsx`, `my_students.json` , `my_students.txt` " are save in your folder automatically .  

## 8. Data Exploration Methods

Before analyzing data, we must first understand the data.

This step is called Data Exploration .

It helps us answer questions like:

How many rows are there?

What are the column names?

Are there missing values?

What type of data is stored?

What is the average or maximum value?

Pandas provides many simple methods to explore data.


**[1] head() - View First Rows**

`head()` - Shows first 5 rows (default)

`df.head(10)` -  Shows first 10 rows

This helps you quickly see how the data looks.

```python
import pandas as pd

data = pd.read_csv('sales_data_sample.csv', encoding = "latin1")

data.head()        # First 5 rows (default)
data.head(10)      # First 10 rows
```

**[2] tail() - View Last Rows**

`tail()` - Shows last 5 rows (default)

`tail(10)` - Shows last 10 rows


```python
import pandas as pd

data = pd.read_csv('sales_data_sample.csv', encoding = "latin1")

data.tail()        # First 5 rows (default)
data.tail(10)      # First 10 rows
```

# **10. Filtering Rows**

Selecting only those rows that match a condition.

For example:

* Show students with marks greater than 80

* Show products with price less than 500

* Show data where category is "Electronics"

* Filtering helps us extract specific data from a large dataset.

Basic Syntax of Filtering
`df[condition]`

If condition is True → Row is selected ,
If condition is False → Row is ignored

Single Condition



```python
import pandas as pd

data = {
    "Name": ["Asha", "Rahul", "Neha", "Rohan"],
    "Marks": [85, 72, 90, 60],
    "City": ["Delhi", "Mumbai", "Delhi", "Pune"]
}

df = pd.DataFrame(data)

# Example: Employees with salary > 50000
filtered_df = df[df["Marks"] > 80]

print(filtered_df)
```

Multiple Conditions (AND - & )

```python
import pandas as pd

data = {
    "Name": ["Asha", "Rahul", "Neha", "Rohan"],
    "Marks": [85, 72, 90, 60],
    "City": ["Delhi", "Mumbai", "Delhi", "Pune"]
}

df = pd.DataFrame(data)

filtered_df = df[(df["Marks"] > 80) & (df["City"] == "Delhi")]
print (filtered_df)



```

Multiple Conditions (OR - | )

```python
import pandas as pd

data = {
    "Name": ["Asha", "Rahul", "Neha", "Rohan"],
    "Marks": [85, 72, 90, 60],
    "City": ["Delhi", "Mumbai", "Delhi", "Pune"]
}

df = pd.DataFrame(data)

filtered_df = df[(df["Marks"] > 80) | (df["City"] == "Pune")]

print (filtered_df)
```
