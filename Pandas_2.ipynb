# **1. Data Modification (Adding, Updating, Removing)**


# **[1.1] Adding New Columns**

✅ Method 1: Direct Assignment

🔹 Syntax:
`df['new_column'] = value`

````python
import pandas as pd

# -----------------------------
# Step 1: Create Sample DataFrame
# -----------------------------

data = {
    'name': ['Ram', 'Shyam', 'Sita'],
    'salary': [50000, 60000, 55000],
    'department': ['HR', 'IT', 'Finance']
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# -----------------------------
# Method 1: Direct Assignment
# -----------------------------

df['bonus'] = df['salary'] * 0.10   # 10% bonus
print("\nAfter Adding 'bonus' Column:")
print(df)
````

✅ Method 2: Insert Method

🔹 Syntax:
`df.insert(position, 'column_name', data)`

````python
# -----------------------------
# Method 2: Insert Method
# -----------------------------


df.insert(0, 'employee_id', [101, 102, 103])
print("\nAfter Inserting 'employee_id' at Position 0:")
print(df)
````

#[1.2] Updating Values

✅ Method 1: Using .loc

🔹 Syntax:
`df.loc[row_index, 'column_name'] = new_value`

````python
# -----------------------------
# Update Single Value using .loc
# -----------------------------

df.loc[0, 'salary'] = 52000   # Update Ram's salary
print("\nAfter Updating Ram's Salary:")
print(df)

````

✅ Method 2: Bulk Update

🔹 Syntax:
`df['column_name'] = df['column_name'] * value:`

````python
# -----------------------------
# Bulk Update (Increase all salaries by 5%)
# -----------------------------

df['salary'] = df['salary'] * 1.05
print("\nAfter Increasing All Salaries by 5%:")
print(df)
````

`df.drop(columns=['col1', 'col2'], inplace=True)`

# **2. Handling Missing Data**


#**[2.1] What is Missing Data?**

When information is not recorded, unavailable, or unknown, it is called missing data.

**How Missing Data Appears in Pandas**

* In Pandas, missing data is usually represented as:

* NaN (Not a Number) → for numeric columns

* None → for object/string columns

````python
import pandas as pd
import numpy as np

# =====================================================
# Step 1: Create Sample Data with Missing Values
# =====================================================

data = {
    'name': ['Ram', 'Shyam', 'Sita', 'Geeta', 'Mohan'],
    'age': [25, np.nan, 30, np.nan, 28],
    'salary': [50000, 60000, np.nan, 55000, np.nan]
}

df = pd.DataFrame(data)

print("Original DataFrame with missinng values:")
print(df)
````

#**[2.2] Detecting Missing Data**

✅ 1. Check Missing Values - It give result in boolean values

 `True` = missing

`False` = value present

🔹 Syntax:
`df.isnull()`



````python
print("\nMissing Value Check (Boolean DataFrame):")
print(df.isnull())
````

✅ 2. Count Missing Values - Counts total missing values in each column

🔹 Syntax:
`df.isnull().sum()`

````python
print("\nCount of Missing Values per Column:")
print(df.isnull().sum())
````

#**[2.3] Handling Missing Data** - There are 3 methods

**✅ Method 1: Drop Missing Data (Delete)**

This method removes missing values completely from the dataset.



🔹 Drop Rows with Missing Values  (rows)

`df.dropna(axis=0, inplace=True)`

**Explanation:**

axis=0 → operates on rows

If a row contains any missing value, it will be removed.

````python
import pandas as pd


# Create sample DataFrame
data = {
    'name': ['Ram', 'Shyam', 'Sita', 'Geeta'],
    'age': [25, np.nan, 30, 28],
    'salary': [50000, 60000, np.nan, 55000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Drop rows with missing values
df.dropna(axis=0, inplace=True)

print("\nAfter Dropping Rows with Missing Values:")
print(df)
````

🔹 Drop Columns with Missing Values (columns)

`df.dropna(axis=1, inplace=True)`

**Explanation:**

axis=1 → operates on columns

If a column contains missing values, the entire column is removed.

````python
import pandas as pd
import numpy as np

# Create sample DataFrame
data = {
    'name': ['Ram', 'Shyam', 'Sita', 'Geeta'],
    'age': [25, np.nan, 30, 28],
    'salary': [50000, 60000, np.nan, 55000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Drop columns with missing values
df.dropna(axis=1, inplace=True)

print("\nAfter Dropping Columns with Missing Values:")
print(df)
````

✅ Method 2: Fill Missing Data (Replace)

Instead of deleting, we replace missing values.

🔹 1️⃣ Fill with a Fixed Value (Example: 0)

✅ Syntax -
`df.fillna(value, inplace=True)`


````python
import pandas as pd
import numpy as np

data = {
    'age': [25, np.nan, 30],
    'salary': [50000, 60000, np.nan]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

df.fillna(0, inplace=True)

print("\nAfter Filling Missing Values with 0:")
print(df)
````

🔹 2️⃣ Fill with Mean (Average)

✅ Syntax - `df['column'].fillna(df['column'].mean(), inplace=True)`

````python
df = pd.DataFrame(data)

df['age'].fillna(df['age'].mean(), inplace=True)
df['salary'].fillna(df['salary'].mean(), inplace=True)

print("\nAfter Filling with Mean:")
print(df)
````

3️⃣ Fill with Median (Alternative for Skewed Data) -

**What is Median?**

Median is the middle value of a sorted dataset.

👉 It is less affected by extreme values (outliers).
That’s why we use median instead of mean when data is skewed.

✅ Syntax -
`df['column'].fillna(df['column'].median(), inplace=True)`

````python
import pandas as pd
import numpy as np

data = {
    'salary': [30000, 35000, 40000, 1000000, np.nan]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Fill missing value with median
df['salary'].fillna(df['salary'].median(), inplace=True)

print("\nAfter Filling Missing Value with Median:")
print(df)
````

🔹 4️⃣ Fill with Mode (For Categorical Data)

📌 What is Mode?

Mode is the value that appears most frequently in a column.

👉 It is commonly used for categorical (string) data like:

* Gender

* City

* Department

* Category

✅ Syntax
`df['column'].fillna(df['column'].mode()[0], inplace=True)`

````python
import pandas as pd
import numpy as np

data = {
    'department': ['HR', 'IT', 'HR', np.nan, 'Finance', 'HR', np.nan]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Fill missing values with Mode
df['department'].fillna(df['department'].mode()[0], inplace=True)

print("\nAfter Filling Missing Values with Mode:")
print(df)
````

 **✅ Method 3: Interpolation (Estimate Middle Value) **- Interpolation is a method used to estimate missing values based on surrounding data.

👉 Instead of deleting or replacing with mean/median,
it calculates a logical value using nearby numbers.

**It is mostly used in:**

* Time-series data

* Temperature records

* Stock prices

* Continuous numeric data

**🔹 1️⃣ Linear Interpolation**

✅ Syntax -
`df['column'].interpolate(method='linear')`

````python
import pandas as pd
import numpy as np

# Create sample data
df = pd.DataFrame({
    'values': [10, np.nan, 30, np.nan, 50]
})

print("Before Interpolation:")
print(df)

# Apply Linear Interpolation
df['values'] = df['values'].interpolate(method='linear')

print("\nAfter Linear Interpolation:")
print(df)
````

# **3. Sorting and Aggregation**


**[3.1] Sorting Data** - Sorting means arranging data in a specific order — either ascending (small → large) or descending (large → small).


````python
import pandas as pd

data = {
    'name': ['Ram', 'Shyam', 'Sita', 'Geeta', 'Mohan'],
    'age': [25, 30, 22, 30, 28],
    'salary': [50000, 60000, 45000, 60000, 52000],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
````

1️⃣ Sorting by a Assending Order

🔹 Syntax -
`df.sort_values(by='column_name')`

````python
df_single_sort = df.sort_values(by='age')

print("\nSorted by Age (Ascending):")
print(df_single_sort)
````

2️⃣ Sorting in Descending Order

🔹 Syntax -
`df.sort_values(by='column_name', ascending=False)`

````python
df_desc_sort = df.sort_values(by='salary', ascending=False)

print("\nSorted by Salary (Descending):")
print(df_desc_sort)
````

✅ 3️⃣ Sorting by Multiple Columns

🔹 Syntax -
`df.sort_values(by=['col1', 'col2'])`

````python
df_multi_sort = df.sort_values(by=['department', 'salary'])

print("\nSorted by Department and Salary (Ascending):")
print(df_multi_sort)
````

4️⃣ Multiple Columns with Different Order

🔹 Syntax
```
df.sort_values(
    by=['col1', 'col2'],
    ascending=[True, False]
```



````python
df_diff_order = df.sort_values(
    by=['department', 'salary'],
    ascending=[True, False]
)

print("\nSorted by Department (Ascending) and Salary (Descending):")
print(df_diff_order)
````

**[3.2] Aggregation Functions** - Aggregation means performing a calculation on a group of data to get a single summary value.

**👉 It answers questions like:**

* What is the average salary?

* What is the total sales?

* What is the maximum age?


````python
import pandas as pd

data = {
    'name': ['Ram', 'Shyam', 'Sita', 'Geeta', 'Mohan'],
    'age': [25, 30, 22, 30, 28],
    'salary': [50000, 60000, 45000, 60000, 52000],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
}

df = pd.DataFrame(data)
print (df)
````

1️⃣ Aggregation on a Single Column

🔹 Syntax: -
`df['column_name'].function()`

````python
# Single column aggregation
print("Total Salary:", df['salary'].sum())
print("Average Salary:", df['salary'].mean())
print("Maximum Age:", df['age'].max())
print("Minimum Age:", df['age'].min())
````

2️⃣ Aggregation on Multiple Columns -

🔹 Syntax: -
`df[['col1', 'col2']].function()`

````python
# Mean of multiple columns
print("Mean of Age and Salary:")
print(df[['age', 'salary']].mean())

# Sum of multiple columns
print("\nSum of Age and Salary:")
print(df[['age', 'salary']].sum())
````

3️⃣ Using agg() for Multiple Functions

🔹 Syntax:
`df['column'].agg(['sum', 'mean', 'max'])`

````python
# Multiple aggregations on salary column
print("Multiple Aggregations on Salary:")
print(df['salary'].agg(['sum', 'mean', 'max', 'min']))
````

4️⃣ Group-wise Aggregation (Very Important) - Used with groupby().

🔹 Syntax: -
`df.groupby('column').agg('function')`

````python
# Average salary department-wise
print("Average Salary by Department:")
print(df.groupby('department')['salary'].mean())

# Multiple aggregations group-wise
print("\nDepartment-wise Salary Stats:")
print(df.groupby('department')['salary'].agg(['sum', 'mean', 'max']))
````

#**4. Grouping Data (GroupBy)**


[4.1] What is Grouping?

Grouping means:

* Splitting data into smaller groups

* Based on common values in a column

👉 It is used when we want to analyze data category-wise.

[4.2] Basic GroupBy Syntax



````python
import pandas as pd

data = {
    'name': ['Ram', 'Shyam', 'Sita', 'Geeta', 'Mohan'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR'],
    'salary': [50000, 60000, 45000, 60000, 52000]
}

df = pd.DataFrame(data)
print (df)
````

1️⃣ Group By Single Column

🔹 Syntax
`df.groupby('column_name')`

````python

# Group by department and calculate average salary
group_single = df.groupby('department')['salary'].mean()

print(group_single)
````

2️⃣ Group By Multiple Columns

🔹 Syntax
`df.groupby(['col1', 'col2'])`

````python
group_multiple = df.groupby(['department', 'salary']).count()

print(group_multiple)
````

[4.3] Applying Aggregation with GroupBy

🔹 What is `groupby()`?

`groupby()` is used to:

Split the data into groups → Apply a function → Combine the results

This process is often called:

🔹 Split → Apply → Combine

**Applying Aggregation with GroupBy**

🔹 General Syntax:
`df.groupby('group_column')['target_column'].function()`

````python
import pandas as pd

data = {
    'name': ['Ram', 'Shyam', 'Sita', 'Geeta', 'Mohan'],
    'age': [25, 30, 22, 30, 28],
    'salary': [50000, 60000, 45000, 60000, 52000],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
}

df = pd.DataFrame(data)
print(df)
````

````python
print("Total Salary by Department:")
print(df.groupby('department')['salary'].sum())
````

2️⃣ mean() – Average Per Group

````python
print("\nAverage Salary by Department:")
print(df.groupby('department')['salary'].mean())
````

3️⃣ count() – Count of Values Per Group

````python
print("\nEmployee Count by Department:")
print(df.groupby('department')['salary'].count())
````

4️⃣ min() – Minimum Value Per Group

````python
print("\nMinimum Salary by Department:")
print(df.groupby('department')['salary'].min())
````

5️⃣ max() – Maximum Value Per Group

````python
print("\nMaximum Salary by Department:")
print(df.groupby('department')['salary'].max())
````

6️⃣ std() – Standard Deviation Per Group

````python
print("\nSalary Standard Deviation by Department:")
print(df.groupby('department')['salary'].std())
````

#**5. Merging and Joining DataFrames**


[5.1] What is Merging?

Merging means combining two DataFrames based on a common column.

It works just like SQL JOINs.

🔹 Basic Syntax:
`pd.merge(left_df, right_df, on='column_name', how='type')`



````python
import pandas as pd

# First DataFrame
employees = pd.DataFrame({
    'emp_id': [1, 2, 3, 4],
    'name': ['Ram', 'Shyam', 'Sita', 'Mohan'],
    'dept_id': [101, 102, 101, 103]
})

# Second DataFrame
departments = pd.DataFrame({
    'dept_id': [101, 102, 104],
    'department_name': ['HR', 'IT', 'Finance']
})

print("Employees DataFrame:")
print(employees)

print("\nDepartments DataFrame:")
print(departments)
````

[5.2] Types of Joins - There are 5 types of joins
- inner join
- outer join
- left join
- right join
- cross join


1️⃣ Inner Join

🔹 Meaning: - Keeps only matching rows from both DataFrames.

🔹 Syntax:-
`pd.merge(df1, df2, on='common_column', how='inner')`


````python
inner_join = pd.merge(employees, departments, on='dept_id', how='inner')
print(inner_join)
````

2️⃣ Left Join

🔹 Meaning: - Keeps all rows from left DataFrame + matching rows from right.

🔹 Syntax: -
`pd.merge(df1, df2, on='column', how='left')`

````python
left_join = pd.merge(employees, departments, on='dept_id', how='left')
print(left_join)
````

3️⃣ Right Join

🔹 Meaning: - Keeps all rows from right DataFrame + matching rows from left.

🔹 Syntax:
`pd.merge(df1, df2, on='column', how='right')`

````python
right_join = pd.merge(employees, departments, on='dept_id', how='right')
print(right_join)
````

4️⃣ Outer Join

🔹 Meaning: - Keeps all rows from both DataFrames.

🔹 Syntax:
`pd.merge(df1, df2, on='column', how='outer')`

````python
outer_join = pd.merge(employees, departments, on='dept_id', how='outer')
print(outer_join)
````

5️⃣ Cross Join

🔹 Meaning: - Creates all possible combinations (Cartesian product).

🔹 Syntax:
`pd.merge(df1, df2, how='cross')`

````python
cross_join = pd.merge(employees, departments, how='cross')
print(cross_join)
````

#**6. Concatenating DataFrames**


Concatenation means:

- Combining two or more DataFrames

- Either vertically (row-wise)

- Or horizontally (column-wise)

Syntax :-
`pd.concat([df1, df2], axis=0, ignore_index=False)`

1️⃣ axis = 0 → Vertical Concatenation (Row-wise)

* DataFrames are stacked one below another

* Rows increase

* Columns must match

🔹 General Syntax - `pd.concat([df1, df2], axis=0)`

````python
import pandas as pd

df1 = pd.DataFrame({
    'name': ['Ram', 'Shyam'],
    'age': [25, 30]
})

df2 = pd.DataFrame({
    'name': ['Sita', 'Mohan'],
    'age': [22, 28]
})

df_vertical = pd.concat([df1, df2], axis=0)

print(df_vertical)
````

2️⃣ axis = 1 → Horizontal Concatenation (Column-wise)

* DataFrames are joined side by side

* Columns increase

* Number of rows should match

🔹 General Syntax - `pd.concat([df1, df2], axis=1)`

````python
df3 = pd.DataFrame({
    'salary': [50000, 60000]
})

df_horizontal = pd.concat([df1, df3], axis=1)

print(df_horizontal)
````

3️⃣ ignore_index Parameter

- Resets index after concatenation

- Creates new index: 0, 1, 2, 3...

🔹 Syntax -
`pd.concat([df1, df2], axis=0, ignore_index=True)`

````python
df_vertical_reset = pd.concat([df1, df2], axis=0, ignore_index=True)

print(df_vertical_reset)
````
